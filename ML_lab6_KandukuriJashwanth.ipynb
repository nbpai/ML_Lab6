```python
#A2- Linear regression
ind=dataset.iloc[:,13]
dep=dataset.iloc[:,12]
ind = ind.dropna()
dep = dep.dropna()
mean_ind=ind.mean()
mean_dep=dep.mean()
# theta-1 calculation
nume=float()
deno=float()
for i,j in zip(ind,dep):
   nume+=(float(i-mean_ind)*float(j-mean_dep))
   deno+=float((i-mean_ind)**2)
print(nume)
print(deno)
theta1=nume/deno
print(theta1)
theta_0=float(mean_dep)-float(theta1*mean_ind)
x=float()
y=theta1*x+theta_0
predic_y=[]
for i in ind:
  value=theta1*i+theta_0
  predic_y.append(value)
print(predic_y)
# mean sum square error
error=float()
for i,j in zip(dep,predic_y):
  error+=((i-j)**2)/len(predic_y)
print(f"Mean sum square error: {error}")



```
